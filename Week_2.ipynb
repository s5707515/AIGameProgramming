```python
#IMPORTS

import math

import matplotlib.pyplot as plt
import numpy as np

#FUNCTIONS
def Sign(_x : float, _theta : float):
  _y = 0.0

  if _x > _theta:

    _y = 1.0

  elif _x < _theta:

    _y = -1.0

  return _y


def Step(_x : float, _theta : float):
  _y = 0.0

  if _x > _theta:

    _y = 1.0

  elif _x < _theta:

    _y = 0.0

  return _y

def Sigmoid(_x : float, _theta : float):
  _y = 0.0

  _y = 1 / (1 + math.e ** -(_x - _theta))

  return _y
```

```python
x = 0.5
y= 0.0
theta = 0.2

x = float(input("Enter the value of X: "))

y = Sign(x, theta)

print("The Sign Function output is: " + str(y))

y = Step(x, theta)

print("The Step Function output is: " + str(y))


y = Sigmoid(x, theta)

print("The Sigmoid Function output is: " + str(round(y,4)) + " (4dp)")

```

## **Line Graphs for functions:**

```python
x_values = np.array([])
y_values = np.array([])


for i in range(-8,8):

  x_values = np.append(x_values, i)
  y_values = np.append(y_values, Sigmoid(i, theta))


plt.plot(x_values, y_values)

plt.xlabel("X Value")
plt.ylabel("Y Value")
plt.title("Sigmoid Function")
plt.show()
```
